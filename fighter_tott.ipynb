```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
fight_results_path = '/mnt/data/ufc_fight_results_processed.csv'
fight_stats_path = '/mnt/data/ufc_fight_stats_processed.csv'

fight_results_df = pd.read_csv(fight_results_path)
fight_stats_df = pd.read_csv(fight_stats_path)

# Merge datasets by EVENT to add weight class information to stats
temp_fight_results = fight_results_df[['EVENT', 'weight_class']]
merged_data_by_event = fight_stats_df.merge(temp_fight_results, on='EVENT', how='left')

# 1. Average Significant Strikes by Weight Class
weight_classes = ['Heavyweight', 'Lightweight']
for weight_class in weight_classes:
    # Filter data by weight class
    weight_class_stats = merged_data_by_event[merged_data_by_event['weight_class'] == weight_class]
    
    # Calculate average significant strikes
    avg_significant_strikes = weight_class_stats['sig_str_land'].mean()
    
    # Plot the result
    plt.figure(figsize=(6, 4))
    plt.bar([weight_class], [avg_significant_strikes], color='blue', edgecolor='black')
    plt.title(f'Average Significant Strikes per Match in {weight_class}', fontsize=14)
    plt.ylabel('Average Significant Strikes', fontsize=12)
    plt.tight_layout()
    plt.show()

# 2. General Distribution of Average Significant Strikes by Round
avg_strikes_by_round = merged_data_by_event.groupby('ROUND')['sig_str_land'].mean()

# Sort rounds logically
avg_strikes_by_round = avg_strikes_by_round.reindex(sorted(avg_strikes_by_round.index))

# Plot strikes by round
plt.figure(figsize=(8, 5))
avg_strikes_by_round.plot(kind='bar', color='teal', edgecolor='black')
plt.title('Average Significant Strikes by Round', fontsize=16)
plt.xlabel('Round', fontsize=12)
plt.ylabel('Average Significant Strikes', fontsize=12)
plt.xticks(rotation=0, fontsize=10)
plt.tight_layout()
plt.show()

# 3. Distribution of Fight Durations (Heavyweight)
heavyweight_fights = merged_data_by_event[merged_data_by_event['weight_class'] == 'Heavyweight']
heavyweight_fights['Duration (Minutes)'] = heavyweight_fights['TIME'] / 60

# Plot duration distribution
plt.figure(figsize=(10, 6))
plt.hist(heavyweight_fights['Duration (Minutes)'], bins=10, color='red', edgecolor='black')
plt.title('Distribution of Fight Durations in Heavyweight', fontsize=16)
plt.xlabel('Duration (Minutes)', fontsize=12)
plt.ylabel('Number of Fights', fontsize=12)
plt.tight_layout()
plt.show()

# 4. Top Heavyweight Fighters by Wins
fighter_wins = pd.concat([
    heavyweight_fights[heavyweight_fights['fighter_1_result'] == 'W']['FIGHTER_1'],
    heavyweight_fights[heavyweight_fights['fighter_2_result'] == 'W']['FIGHTER_2']
]).value_counts()

# Select top 10
top_heavyweight_fighters = fighter_wins.head(10)

# Plot top fighters by wins
plt.figure(figsize=(12, 6))
top_heavyweight_fighters.plot(kind='bar', color='green', edgecolor='black')
plt.title('Top Heavyweight Fighters by Wins', fontsize=16)
plt.xlabel('Fighter', fontsize=12)
plt.ylabel('Number of Wins', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```
